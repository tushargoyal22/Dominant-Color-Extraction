```python
X,y = make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

## K-Means is an unsupervised learning algorithm , that said we dont give y values to algorithm

```python
k=5

color = ["green","red","blue","yellow","orange"]

clusters = {}
for i in range(k):
    center = 10* (2*np.random.random((X.shape[1],))-1) # its in -10 to 10  , as np.random.random gives in 0 to 1 so multiplying by 2 -> 0 to 2 then -1 to 1 then -10 to 10
    points= []
    cluster ={
        'center':center,
        'points':points,
        'color':color[i]
    }
    
    clusters[i]=cluster
```

```python
#clusters
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
v1 = np.array([1,2])
v2 = np.array([3,5])

distance(v1,v2)
```

```python
#Assign every Point to One of clusters
def assignpointToClusters(clusters): #E-Step
    for ix in range(X.shape[0]):
        dist = []
        curr_x = X[ix]

        for kx in range(k):
            d = distance(curr_x,clusters[kx]['center'])
            dist.append(d)
        current_cluster  = np.argmin(dist)

        clusters[current_cluster]['points'].append(curr_x)

        
def updateClusters(clusters):# M-Step --> We update  every cluster center according to mean of cluster
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        if(pts.shape[0]>0):# If a cluster has some non zero points
            new_u =pts.mean(axis=0)
            clusters[kx]['center']=new_u
            clusters[kx]['points'] = []# clear the assigned points to cluster as center we have updated so we will re assign the points
        else:
            clusters[kx]['center'] = 10* (2*np.random.random((X.shape[1],))-1)
            clusters[kx]['points'] = []# clear the assigned points to cluster as center we have updated so we will re assign the points
            print("Center was very far from data points , defining this center randomly again")
            
def plotCluster(clusters):
    
    for kx in range(k):
        pts= np.array(clusters[kx]['points'])
#         print(pts.shape)
        # Plot the points 
        try:
            
            plt.scatter(pts[:,0],pts[:,1])
#             plt.show()
        
        except:
            pass
        
        #Plot the cluster points
        
        uk=clusters[kx]['center']
        plt.scatter(uk[0],uk[1],color="black",marker="*")
        
```

The plt.show() command does a lot under the hood, as it must interact with your systems interactive graphical backend. The details of this operation can vary greatly from system to system and even installation to installation, but Matplotlib does its best to hide all these details from you.

One thing to be aware of: the plt.show() command should be used only once per Python session, and is most often seen at the very end of the script. Multiple show() commands can lead to unpredictable backend-dependent behavior, and should mostly be avoided.

- E step basically populates the list of points to a cluster
- M step updates the cluster centers according to the populated cluster
